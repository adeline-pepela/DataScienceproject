## MSC DSA Scholars Exam Data


 Importing the necessary libraries

```python
import pandas as pd #read csv
import numpy as np #read arrays
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# loading training data
tr_data = pd.read_csv("Train_data.csv")
print("\ntraining data", tr_data.head())

# loading validation data
val_data = pd.read_csv("Validation.csv")
print("\nvalidation data", val_data.head())
```

```python
# To display the sum of null values of the datasets
print("\ntraining data", tr_data.isnull().sum())

print("\nvalidation data", val_data.isnull().sum())
```

```python
# replacing the null values using mean
df_tr = tr_data.fillna(tr_data.mean(numeric_only = True))

df_val = val_data.fillna(val_data.mean(numeric_only = True))
```

```python
# To verify the sum of null values of the datasets
print("\ntraining data", df_tr.isnull().sum())

print("\nvalidation data", df_val.isnull().sum())
```

```python
# Visualizing the distribution of well being index
plt.figure(figsize=(10, 8))
sns.histplot(df_tr["Well-Being Index"], kde=True)
plt.title("Distribution of Well being index")
plt.xlabel("Well Being index")
plt.ylabel("Frequency")
plt.show()
```

The histogram visualization of the well-being index reveals a right-skewed distribution, indicating a notable trend within the galaxies' well-being index values. Specifically, this skewness implies a higher occurrence of galaxies with lower well-being index values, while fewer galaxies exhibit higher values.

In a right-skewed distribution, the tail of the distribution stretches towards the higher values, while the majority of data clusters towards the lower end of the scale. Applied to the well-being index, this observation suggests that a larger proportion of galaxies experience relatively lower levels of well-being compared to those with higher levels of well-being.
